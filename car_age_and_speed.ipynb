## A PYTHON PROGRAM THAT SHOWS VEHICLES AGES AND THEIR SPEEN IN (KM/H)
### USING SCATTER PLOTS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91])
sizes = np.array([190,190,105,200,300,100,200,100,150,1010, 200,100, 109])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.xlabel("CAR AGE")
plt.ylabel("CAR SPEED (KM/H)")
plt.style.use("fivethirtyeight")
plt.title("A SCATTER PLOT SHOWING CAR AGES AND THEIR SPEED IN (KM/H)", fontsize = "15", color = "brown")

plt.grid(True)
plt.scatter(x, y, c=colors, s=sizes)


plt.show()
```
